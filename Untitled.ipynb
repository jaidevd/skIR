```python
import elastic
import pandas as pd
from spacy import load
from elasticsearch_dsl import Index
from elasticsearch_dsl import field
import numpy as np
from elasticsearch_dsl.connections import create_connection
import en_core_web_lg
```

```python
df=pd.read_csv('APY_1997-98_to_2015-16.csv') #reading in the csv file
conn = create_connection(hosts=['localhost'])
df1 = df.replace(np.nan, '', regex=True) #taking care of the NaN values
elastic.index_table(df1, 'agriculture1', conn) #creating an index with entries from df
```

```python
nlp=en_core_web_lg.load()
index =Index('agriculture1')
search_engine = elastic.TableSearch(nlp, index)  #creating a search instance for index specified above
```

```python
hits=search_engine.query('Production of khafir in 2007-08').hits #query which returns multiple responses
```

```python
for response in hits:   #interating through the responses
    for key in response:     #iterating through the fields in every response
        print(key+":"+str(response[key]))
    print("\n")
```
