```python
import elastic
```

```python
import pandas as pd
from spacy import load
from elasticsearch_dsl import Index
from elasticsearch_dsl import field
import numpy as np
from elasticsearch_dsl.connections import create_connection
import en_core_web_lg
```

```python
df=pd.read_csv('APY_1997-98_to_2015-16.csv')
conn = create_connection(hosts=['localhost'])
df1 = df.replace(np.nan, '', regex=True)
elastic.index_table(df1, 'agriculture1', conn)
```

```python
nlp=en_core_web_lg.load()
index = Index('agriculture1')
search_engine = elastic.TableSearch(nlp, index)
hits=search_engine.query('Production of khafir in 2007-08').hits
```

```python
for response in hits:
    for key in response:
        print(key+":"+str(response[key]))
    print("\n")
```
